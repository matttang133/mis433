## Question 4. Adding new columns

Currently dep_time and sched_dep_time are convenient to look at, but hard to compute with because they’re not really continuous numbers.

For example, 759 means 7:59 and 801 means 8:01. Their difference is not 42 but 2 minutes.

```python
# Convert them to a more convenient representation of number of minutes since midnight (0).

def time_to_minutes(time_val):
    if pd.isna(time_val):
        return pd.NA
    time_val = int(time_val)
    hours = time_val // 100
    minutes = time_val % 100
    return hours * 60 + minutes

flights['dep_time_minutes'] = flights['dep_time'].apply(time_to_minutes)
flights['sched_dep_time_minutes'] = flights['sched_dep_time'].apply(time_to_minutes)

display(flights[['dep_time', 'dep_time_minutes', 'sched_dep_time', 'sched_dep_time_minutes']].head())
```

```python
# Create a new column of arr_time - dep_time.
# Convert arr_time to minutes since midnight for consistent calculation
flights['arr_time_minutes'] = flights['arr_time'].apply(time_to_minutes)

# Calculate flight duration, accounting for overnight flights
def calculate_flight_duration(dep_min, arr_min):
    if pd.isna(dep_min) or pd.isna(arr_min):
        return pd.NA
    if arr_min < dep_min:
        return (arr_min + 24 * 60) - dep_min
    return arr_min - dep_min

flights['calculated_flight_duration'] = flights.apply(
    lambda row: calculate_flight_duration(row['dep_time_minutes'], row['arr_time_minutes']),
    axis=1
)

# Compare this column with air_time.
display(flights[['air_time', 'calculated_flight_duration']].head())

flights['duration_diff'] = flights['calculated_flight_duration'] - flights['air_time']
print("\nDifference between calculated duration and air_time (first 5 rows):")
display(flights[['air_time', 'calculated_flight_duration', 'duration_diff']].head())
print("\nSummary of differences between calculated duration and air_time:")
display(flights['duration_diff'].describe())
```

```python
# Compare dep_time, sched_dep_time, and dep_delay. How would you expect those three numbers to be related?
# Try creating a column to calculate dep_delay from dep_time and sched_dep_time (and/or other columns if necessary).
flights['calculated_dep_delay'] = flights['dep_time_minutes'] - flights['sched_dep_time_minutes']

# Test your results.
print("Comparison of existing 'dep_delay' and 'calculated_dep_delay' (first 5 rows):")
display(flights[['dep_delay', 'calculated_dep_delay']].head())

consistency_check = (flights['dep_delay'] == flights['calculated_dep_delay'])
print("\nConsistency check between 'dep_delay' and 'calculated_dep_delay':")
display(consistency_check.value_counts(dropna=False))

# To find mismatched rows, we need to handle pd.NA in the consistency_check Series.
# We treat rows where consistency_check is False or pd.NA as mismatches.
mismatched_delays = flights[consistency_check.fillna(False) == False]
if not mismatched_delays.empty:
    print("\nMismatched rows between 'dep_delay' and 'calculated_dep_delay':")
    display(mismatched_delays[['dep_time', 'sched_dep_time', 'dep_delay', 'calculated_dep_delay']].head())
```
